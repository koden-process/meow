```python
def rename_attributes(doc, rename_map):
    attributes = doc.get("attributes", {})
    new_attributes = {}

    for key, value in attributes.items():
        new_key = rename_map.get(key, key)  # si pas dans mapping → on garde la clé
        new_attributes[new_key] = value

    doc["attributes"] = new_attributes
    return doc
```

```python
def rewrite_references(doc, rename_map):
    if "references" in doc:
        for ref in doc["references"]:
            old_key = ref.get("schemaAttributeKey")
            if old_key in rename_map:
                ref["schemaAttributeKey"] = rename_map[old_key]

    return doc
```

## Events
- changement de teamId
- mappage des lanes dans body

```python
print("# changement de teamId")
result = db["Events"].update_many(
    {},
    {'$set': {'teamId': ObjectId(teamId)}}
)
print(f"matched={result.matched_count}, modified={result.modified_count}")

print("# mappage des lanes dans body")
for lane, old_lanes in lanes.items():
    print(f"# Transfert vers la lane {lane}")
    for old_lane in old_lanes:
        result = db["Events"].update_many(
            {'body.from': old_lane},
            {'$set': {
                'body.from': lane
            }}
        )
        print(f"body.from ({old_lane}) : matched={result.matched_count}, modified={result.modified_count}")

    for old_lane in old_lanes:
        result = db["Events"].update_many(
            {'body.to': old_lane},
            {'$set': {
                'body.to': lane
            }}
        )
        print(f"body.to ({old_lane}) : matched={result.matched_count}, modified={result.modified_count}")
```
